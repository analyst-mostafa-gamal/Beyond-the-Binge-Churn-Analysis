```python
# Selecting all columns except the subscriber_id column
cols_to_check = [col for col in df.columns if col != 'subscriber_id']
print(cols_to_check)
```

```python
# Checking number of duplicate values on subset
df.duplicated(subset= cols_to_check).sum()
```

```python
# Removing duplicates based on subset and keep only the first occurrence
df = df.drop_duplicates(subset= cols_to_check, keep='first')
```

```python
# Checking number of duplicate values after removing duplicates based on subset
df.duplicated(subset= cols_to_check).sum()
```

```python
# Converting monthly_spend_eur to numeric
df['monthly_spend_eur'] = pd.to_numeric(df['monthly_spend_eur'], errors='coerce')
```

```python
# Selecting float and int columns only
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Checking for negative values in the specified columns
for col in num_cols:
    if col in df.columns:
        neg_count = len(df[df[col] < 0])
        if neg_count > 0:
            print(f"⚠️ col '{col}': there are {neg_count} negative values")
        else:
            print(f"✅ col '{col}': clean (no negative values)")
```

```python
# Selecting text columns only
text_cols = df.select_dtypes(include=['object']).columns

# Checking hidden spaces (extra spaces)
for col in text_cols:
    has_spaces = df[col].astype(str).str.contains(r'^\s|\s$', regex=True).sum() # counts how many values in column [col] have spaces at the beginning or end

    if has_spaces > 0:
        print(f"⚠️ col [{col}]: there are {has_spaces} rows with extra spaces")
    else:
        print(f"✅ col [{col}]: clean (no extra spaces)")
```

```python
# Replaceing unrealistic age values (less than 15 or greater than 85) with NaN (missing value)
df.loc[(df['age'] < 15) | (df['age'] > 85), 'age'] = np.nan

# Filling missing values with Median
df.loc[:, 'age'] = df['age'].fillna(df['age'].median())

# Changing the data type of the age column to integer
df['age'] = df['age'].astype(int)

# Grouping age into age_groups
def age_bins(age):
    if age < 25:
      return '1. Gen Z (<25)'
    elif age < 40:
      return '2. Millennials (25-39)'
    elif age < 60:
      return '3. Gen X (40-59)'
    else:
       return '4. Boomers (60+)'

# Creating a new column 'age_group' in the DataFrame 'df'
df['age_group'] = df['age'].apply(age_bins)

# Checking the result (sample display)
print(df[['age', 'age_group']].sample(10))

# Checking the value counts of the 'age_group' column to see the distribution of age groups in the dataset
df['age_group'].value_counts().sort_index()
```

```python
# Replacing any negative values in the subscription_tenure_months column with 0.0
df.loc[df['subscription_tenure_months'] < 0.0, 'subscription_tenure_months'] = 0.0

# Type casting (implicit type conversion)
df['subscription_tenure_months'] = df['subscription_tenure_months'] + 0.0 # Removing the (Negative Zero)

# Rounding subscription_tenure_months to 2 decimal places
df['subscription_tenure_months'] = df['subscription_tenure_months'].round(2)

# Gruoping subscription_tenure_months into tenure groups
def categorize_tenure(months): # Definition of the classification function
    if months < 1.0:
      return '1. New User (<1 Month)'
    elif months <= 6.0:
      return '2. Short Term (1-6 Months)'
    elif months <= 12.0:
      return '3. Mid Term (6-12 Months)'
    elif months <= 24.0:
      return '4. Long Term (1-2 Years)'
    else:
      return '5. Loyal (2+ Years)'

# Creating a new column 'tenure_group' in the DataFrame 'df'
df['tenure_group'] = df['subscription_tenure_months'].apply(categorize_tenure)

# Checking the result (sample display)
print(df[['subscription_tenure_months', 'tenure_group']].sample(10))

# Checking the value counts of the 'tenure_group' column to see the distribution of tenure groups in the dataset
df['tenure_group'].value_counts().sort_index()
```

```python
# Rounding support_tickets_raised to the nearest integer
df['support_tickets_raised'] = df['support_tickets_raised'].round(0).astype(int)

# Grouping support_tickets_raised into support_groups
def classify_support(tickets): # Definition of the classification function
    if tickets == 0:
      return '0. No Issues'
    elif tickets <= 2:
      return '1. Low Support Needs'
    else:
      return '2. High Support Needs (Potential Churn)'

# Creating a new column 'support_priority' in the DataFrame 'df'
df['support_priority'] = df['support_tickets_raised'].apply(classify_support)

# Checking the result (sample display)
print(df[['support_tickets_raised', 'support_priority']].sample(10))

# Checking the value counts of the 'support_priority' column to see the distribution of support priorities in the dataset
df['support_priority'].value_counts().sort_index()
```

```python
# Making sure there are no negative values ​​(the minimum is 0) and Handling extreme values ​​(Outliers)
df['satisfaction_score'] = df['satisfaction_score'].clip(lower=0, upper=10)

# Filling missing values with Median
df['satisfaction_score'] = df['satisfaction_score'].fillna(df['satisfaction_score'].median())

# Rounding satisfaction_score to the nearest integer
df['satisfaction_score'] = df['satisfaction_score'].round(0).astype(int)

# Grouping satisfaction_score into satisfaction groups
def categorize_satisfaction(score):
    if score <= 3:
      return '1. Detractors (Unsatisfied)'
    elif score <= 7:
      return '2. Passives (Neutral)'
    else:
      return '3. Promoters'

# Creating a new column 'satisfaction_group' in the DataFrame 'df'
df['satisfaction_group'] = df['satisfaction_score'].apply(categorize_satisfaction)

# Checking the result (sample display)
print(df[['satisfaction_score', 'satisfaction_group']].sample(10))

# Checking the value counts of the 'satisfaction_score' column to see the distribution of satisfaction scores in the dataset
df['satisfaction_group'].value_counts().sort_index()
```

```python
# Making sure there are no negative values ​​(the minimum is 0) and Handling extreme values ​​(Outliers)
df['promo_discount_pct'] = df['promo_discount_pct'].clip(lower=0.0, upper=30.0) # Assuming that the maximum reasonable discount is 30

# Rounding promo_discount_pct to 2 decimal places
df['promo_discount_pct'] = df['promo_discount_pct'].round(2)

# Grouping promo_discount_pct into discount groups
def categorize_discount(pct):
  if pct <= 5.0:
    return '1. Starter (0-5%)'
  elif pct <= 15.0:
    return '2. Silver (5-15%)'
  else:
    return '3. Gold (>15%)'

# Creating a new column 'discount_group' in the DataFrame 'df'
df['discount_group'] = df['promo_discount_pct'].apply(categorize_discount)

# Checking the result (sample display)
print(df[['promo_discount_pct', 'discount_group']].sample(10))

# Checking the value counts of the 'discount_group' column to see the distribution of discount groups in the dataset
df['discount_group'].value_counts().sort_index()
```

```python
# Filling missing values in the 'days_since_last_activity' column with 999 to mark them forever since last activity
df['days_since_last_activity'] = df['days_since_last_activity'].fillna(999.0)

# Type casting (implicit type conversion)
df['days_since_last_activity'] = df['days_since_last_activity'] + 0.0 # Removing the (Negative Zero)

# Rounding days_since_last_activity to 2 decimal places
df['days_since_last_activity'] = df['days_since_last_activity'].round(2)

# Groupings days_since_last_activity into activity groups
def categorize_activity(days):
  if days == 999.0:
    return '0. Never Active'
  elif days <= 1.0:
    return '1. Very Active (Today)'
  elif days <= 7.0:
    return '2. Active (This Week)'
  elif days <= 30.0:
    return '3. Slipping (Past Month)'
  elif days <= 90.0:
    return '4. At Risk (1-3 Months)'
  else:
    return '5. Inactive (>3 Months)'

# Creating a new column 'activity_status' in the DataFrame 'df'
df['activity_status'] = df['days_since_last_activity'].apply(categorize_activity)

# Checking the result (sample display)
print(df[['days_since_last_activity', 'activity_status']].sample(10))

# Checing the value counts of the 'activity_status' column to see the distribution of activity statuses in the dataset
df['activity_status'].value_counts().sort_index()
```

```python
# Replacing any negative values in the avg_bet_amount_eur column with 0
df.loc[df['avg_bet_amount_eur'] < 0.0, 'avg_bet_amount_eur'] = 0.0

# Type casting (implicit type conversion)
df['avg_bet_amount_eur'] = df['avg_bet_amount_eur'] + 0.0 # Removeing the (Negative Zero)

# Rounding avg_bet_amount_eur to 2 decimal places
df['avg_bet_amount_eur'] = df['avg_bet_amount_eur'].round(2)

# Groupings avg_bet_amount_eur into bet_groups
def get_monetary_segment(amount):
  if amount == 0.0:
    return '0. Non-Spender'
  elif amount <= 5.0:
    return '1. Micro Bettor (<5 eur)'
  elif amount <= 50.0:
    return '2. Standard (5-50 eur)'
  elif amount <= 200.0:
    return '4. High Roller (50-200 eur)'
  else:
    return '5. Whale / VIP (>200 eur)'

# Creating a new column 'customer_value_segment' in the DataFrame 'df'
df['customer_value_segment'] = df['avg_bet_amount_eur'].apply(get_monetary_segment)

# Checking the result (sample display)
print(df[['avg_bet_amount_eur', 'customer_value_segment']].sample(10))

# Checking the value counts of the 'customer_value_segment' column to see the distribution of customer value segments in the dataset
df['customer_value_segment'].value_counts().sort_index()
```
